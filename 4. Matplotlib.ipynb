```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore") #Hides futures warnings
```

```python
#LINE CHART
years = range(2008,2020)
demand = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
supply = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
# First line
plt.plot(years, demand, 'b-x', linewidth=2, markeredgecolor='purple', markersize=8, markeredgewidth=3, color = 'green');

# Second line
plt.plot(years, supply, 'r--o', linewidth=2, markersize=8, markeredgecolor ='orange',color = 'red');

# Title
plt.title('Demand vs supply over the years');

# Line labels
plt.legend(['Demand', 'Supply'])

# Axis labels
plt.xlabel('Year'); plt.ylabel('Quantity');
```

Lets make a dataframe to plot using numpy and pandas

```python
df = pd.DataFrame(np.random.randn(100,3))
df['year'] = range(1920,2020)
cols = ['year',0,1,2]
df = df[cols].copy()
```

```python
df
```

```python
datafig, axes = plt.subplots(1,3, figsize = (15,4))
plt.tight_layout(pad =2)

axes[0].plot(df.year,df[0], 'o--b')
axes[0].set_title('Data:0');
axes[0].legend(['Numbers'])
axes[0].set_xlabel('Digits')
axes[0].set_ylabel('Random Numbers');


axes[1].scatter(df[1],df[2])
axes[1].scatter(df[0],df[2])
axes[1].scatter(df[1],df[0])
axes[1].set_title('Data:1');
axes[1].legend(['Numbers'])
axes[1].set_xlabel('Digits')
axes[1].set_ylabel('Random Numbers')


axes[2].boxplot(df.iloc[:,1:4]) #Also see bar plot
axes[2].set_title('Data:2');
axes[2].legend(['Numbers'])
axes[2].set_xlabel('Digits')
axes[2].set_ylabel('Random Numbers');
```

```python
fig, ax1 = plt.subplots(figsize=(30,12), dpi=100)
data1=df; ax1.plot(df.year, df[0],color='g',linewidth=2, linestyle='--', marker='o', markerfacecolor='g', markersize=12)
ax1.set_ylabel("Data: 0", color='g', fontsize=25); ax1.text(1990, 2.5, 'r=-0.0035*', style='italic', fontsize=25, color = 'g')
ax1.axhline(y=0, color='k', linestyle='-.',linewidth=2 )
ax1.set_ylim(-3, 3); ax1.set_xlim(1980, 2019)
ax1.set_xlabel("Years", fontsize=25);

ax2 = ax1.twinx()
color = 'r'
ax2.plot(df.year, stats.zscore(df[1]),color='r', linewidth=2, linestyle='-.', marker='s', markerfacecolor='r', markersize=12, label="PMM Index")
ax2.set_ylabel("Data: 1", color='r',fontsize=30); ax2.tick_params(axis='y', labelcolor=color, labelsize=25)
ax2.set_ylim(-3, 3);


ax3 = ax1.twinx();
ax3.spines['right'].set_position(('outward', 100))
color = 'b'
ax3.plot(df.year,stats.zscore(df[2]),color='b', linewidth=2, linestyle='-', marker='*', markerfacecolor='b', markersize=12, label="NAO Index")
ax3.set_ylabel("Data: 2", color='b',fontsize=30); ax3.tick_params(axis='y', labelcolor=color, labelsize=25)
ax3.set_ylim(-3, 3);
```

We can see a few styles in seaborn which is built on matplotlib and blends well with matplotlib codes for labeling, title, etc.:

```python
import seaborn as sns
import matplotlib
```

```python
# Configuring styles 
sns.set_style("whitegrid")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#FFFFFF'
```

```python
#Using seaborn to plot from a famous inbuilt dataset of petal and sepal lengths and widths 
#with 150 rows and 5 columns
data = sns.load_dataset("iris")
# View the data
print(data.shape)
data.sample(5)
```

```python
# Scatterplot with seaborn
sns.scatterplot(x = data.sepal_length, 
                y = data.sepal_width, 
                hue=data.species, 
                s=100);
# Chart title
plt.title("Flowers");
```

```python
plt.title('Distribution plot for sepal widths')
sns.distplot(data.sepal_width);
```

```python
sns.boxplot(x="variable", y="value", data=pd.melt(data.iloc[:,:-1]))
```

```python
sns.barplot(x="variable", y="value", data=pd.melt(data.iloc[:,:-1]));
```

```python
sns.lineplot(data=data.sepal_length, color=u'#dd0101', alpha = 0.7).set_ylabel('Sepal length', color=u'#dd0101')
ax2 = plt.twinx()
sns.lineplot(data=data.sepal_width, color=u'#2e00ff', ax=ax2, alpha = 0.7)
ax2.lines[0].set_linestyle("--")
ax2.set_ylabel('Sepal width', color=u'#2e00ff');
# ax3 = plt.twinx()
# sns.lineplot(data=data.petal_length, color="r", ax=ax2)
```
