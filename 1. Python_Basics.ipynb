```python
##Functions 

#To find the remainder in division

def remainder():
    print('please provide a integer number: ')
    x = int(input())
    print('please provide a divisor: ')
    y = int(input())
    rem = x % y
    print('The remainder is: ')
    return rem
```

```python
remainder()
```

```python
#Note division, remainder etc
print(33/2) #Gives float value
print(33%2) #Gives remainder

print(33//2) #gives int value -- also called floor division 
```

```python
#Lambda function -- simple one line function for ex which may be used once 
#lambda arguments: expression
#Calculting cube of a number using function and lambda function:

def cube(y):
    return y*y*y
 
lambda_cube = lambda y: y*y*y

print(cube(3))
print(lambda_cube(3))
```

```python
rt = [1, 4, 5, 2, 3]
rt*2  #repeats twice 
```

```python
#Using lambda and map 
rt = [1, 4, 5, 2, 3]
 
listtwice = list(map(lambda x: x*2, rt))
print(listtwice)
```

```python
#The base python does not have functions such as square root, pi, e etc. But you can always import lots of packages
# or modules for it which have built in functions 
import math

r = 3

area = math.pi * r * r
print(area)

print(math.sqrt(9))
#Shift+Tab to find help on a function
print(math.pow(3,3))
print(int(math.pow(27,1/3)))
```
